```python
# 1. Humanクラスを定義
class Human:
    # 初期化メソッド（属性 name と age を持つ）
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # 名前と年齢を出力するメソッドを追加
    def printinfo(self):
        print(f"名前: {self.name}")
        print(f"年齢: {self.age}")

# 2. Humanクラスのインスタンスを作成し、変数に代入
person = Human("東村幸太郎", 19)

# 3. メソッド(printinfo)を呼び出して出力
person.printinfo()
```
