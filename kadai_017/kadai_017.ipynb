```python
# 【クラス】Humanクラスの定義
class Human:
    # 【関数】【変数のスコープ】コンストラクタでインスタンス変数（属性）を初期化
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # 【関数】【条件分岐】大人かどうかをチェックして出力するメソッド
    def check_adult(self):
        # ageが20以上かどうかの条件分岐
        if self.age >= 20:
            print(f"{self.name}は大人です。（年齢: {self.age}歳）")
        else:
            print(f"{self.name}は大人ではありません。（年齢: {self.age}歳）")

# 【リスト】Humanクラスのインスタンスを複数生成し、リストに追加
humans = [
    Human("侍太郎", 25),
    Human("侍次郎", 18),
    Human("侍花子", 20),
    Human("侍三郎", 12)
]

# 【繰り返し処理】リストから要素（インスタンス）を1つずつ取り出し、メソッドを呼び出す
for human in humans:
    human.check_adult()
```
