```python
# 1. 消費税を加えた計算結果を返す関数を定義
def calculate_tax(price, tax_rate):
    # 消費税額 = 金額 × (税率 / 100)
    # 合計金額 = 金額 + 消費税額
    total_price = price + (price * (tax_rate / 100))

    # 3. 戻り値で計算結果を返す
    return total_price

# 2. 定義した関数を呼び出し、4. 引数を設定
result = calculate_tax(110, 10)

# 結果を出力
print(result)
```
